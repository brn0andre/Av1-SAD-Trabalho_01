```python
# Importando bibliotecas

import pandas as pd
import numpy as np+
import os
```

```python
# Criando variáveis de cada caminho de diretório

# Arquivo de origem com os dados brutos da PRF de 2025
caminho_raw = r"C:\Users\andre\OneDrive\SAD\Trabalho_Transito\data\raw\datatran2025.csv"

# Diretório onde serão salvos os arquivos após a limpeza e transformação
pasta_processed = r"C:\Users\andre\OneDrive\SAD\Trabalho_Transito\data\processed"
```

```python
# Criando pasta processed

os.makedirs(pasta_processed,exist_ok=True)
```

```python
# Leitura dos dados
# Carregando arquivo CSV utilizando separador "," 
# Carregando com encoding latin-1 por conta dos acentos e caracteres especiais

df = pd.read_csv(caminho_raw, sep=",", encoding="latin-1")

# Verificando as primeiras linhas e a estrurura das colunas
print(df.info())
```

```python
# Salvando em processed (UTF-8)

# Criando caminho onde vamos alocar o arquivo csv que está carregado na variável df
caminho_processed = os.path.join(pasta_processed, "datatran2025.csv")

# Carregando o arquivo csv presente na variável df para pasta processed
df.to_csv(caminho_processed, index=False, encoding="utf-8")

print("Arquivo carregado e salvo com sucesso 🚀")
```
