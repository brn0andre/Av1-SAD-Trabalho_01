#### Importando Bibliotecas

```python
# Importando bibliotecas

import pandas as pd
import numpy as np
```

```python
# Carregando dataset tratado
df = pd.read_csv("../data/processed/datatran2025.csv")
```

##### Análise Exploratória de Dados(EDA) :

```python
# Verificando os dados em cada coluna para achar valores nulos 
# Além de verificar colunas que não serão relevantes para a análise.

print("--- Primeiras Linhas ---")
display(df.head(5)) # O display() deixa a tabela mais bonita no Jupyter que o print()

print("\n--- Lista de Colunas ---")
print(df.columns)

print("\n--- Resumo Estatístico (Dados Numéricos e Categóricos) ---")
display(df.describe(include="all"))

print("\n--- Contagem de Valores Nulos (NaN) ---")
print(df.isnull().sum())
```

```python
# Selecionando as colunas relevantes para a Análise
colunas_selecionadas = [
    'uf',
    'fase_dia',
    'tipo_pista',
    'causa_acidente',
    'tipo_acidente',
    'condicao_metereologica',
    'mortos',
    'feridos'
]

df_modelo = df[colunas_selecionadas].copy()


# Criando a variável Alvo (Target)
# 0 = Sem vítimas
# 1 = Com vítimas (feridos ou mortos)
# Recebe 1 se houver mortos ou feridos, caso contrário 0.
# A condição gera valores booleanos (True/False), que são convertidos para inteiro com astype(int).

df_modelo['target'] = (
    (df_modelo['mortos'] > 0) | (df_modelo['feridos'] > 0)
).astype(int)

# Convertendo em int pois ele por padrão retorna true or false
```

```python
display(df_modelo)
```

```python
# Verificação da quantidade de Valores Nulos Existentes
print(df_modelo.isnull().sum())

# Tratamento de nulos
df_modelo = df_modelo.dropna()

# Mostra 10 linhas sorteadas aleatoriamente
display(df_modelo.sample(10))
```

```python
# Convertendo as colunas de contagem para inteiros pois após utilizarmos o tratamento de valores nulos...
# O pandas converte toda a coluna para float.

df_modelo['mortos'] = df_modelo['mortos'].astype(int)
df_modelo['feridos'] = df_modelo['feridos'].astype(int)

# 
df_modelo['target'] = df_modelo['target'].astype(int)

display(df_modelo.head())
```

```python
# Salvando o resultado do tratamento em um novo arquivo
# index=False evita que o Pandas crie uma coluna extra de números sem nome
df_modelo.to_csv('../data/processed/dados_limpos_para_modelo.csv', index=False)

print("Arquivo salvo com sucesso!")
```
